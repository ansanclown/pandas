## 9.데이터선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df['키']>=185 #학생들 키가 185이상인지 true, false
```

```python
filt = (df['키']>=185)
df[filt]
```

```python
df[~filt] #filter 역으로
```

```python
df[df['키']>=185]
```

```python
#df.loc[row,col]
df.loc[df['키']>=185,'수학'] #키 185 이상인 학생의 수학 데이터
```

```python
df.loc[df['키']>=185,['이름','수학','과학']]
```

## 다양한 조건

### &(그리고)

```python
df.loc[(df['키']>=185) & (df['학교'] == '북산고')] #키가 185 & 북산고
```

## |(또는)

```python
df.loc[(df['키'] < 170) | (df['키']>200)]
```

## str 함수

```python
filt = df['이름'].str.startswith('송') #송씨 성
df[filt]
```

```python
filt = df['이름'].str.contains('태') #태가 들어가는 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') #태가 들어가는 사람
df[~filt] #이름에 태가 들어가는 사람 제외
```

```python
langs = ['Python','Java']
filt = df['SW특기'].isin(langs) #SW 특기가 파이썬이거나 자바인사람
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs) #SW 특기가 파이썬이거나 자바인사람
df[filt]
```

```python
filt=df['SW특기'].str.contains('java')
df[filt]
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java',na=True)
```

```python
df['SW특기'].str.contains('Java',na=False) #NaN에 대해 False
```

```python
filt=df['SW특기'].str.contains('Java',na=False)
df[filt]
```

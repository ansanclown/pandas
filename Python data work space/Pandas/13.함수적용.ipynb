## 데이터에 함수 적용 (apply)

```python
#키 뒤에 cm
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) #키 데이터에 대해 add_cm을 호출한 결과 데이터를 반영
df
```

```python
def cap(lang):
    if pd.notnull(lang): # NaN이 아닌지
        return lang.capitalize() #첫글자는 대문자
    return lang

df['SW특기'] = df['SW특기'].apply(cap)
df
```

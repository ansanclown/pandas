# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 데이터 처리

```python
import pandas as pd
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').get_group('능남고')
```

```python
df.groupby('학교').mean(numeric_only=True) #계산 가능한 데이터 평균
```

```python
df.groupby('학교').size() # 각 그룹의 크기정보
```

```python
df.groupby('학교').size()['능남고'] #학교로 그룹화 한뒤 능남고에 해당하는 데이터 수
```

```python
df.groupby('학교')['키'].mean() #학교로 그룹화 한뒤 키 평균
```

```python
df.groupby('학교')[['국어','영어','수학']].mean() #학교로 그룹화 한 뒤 국영수 데이터
```

```python
df['학년'] = [3,3,2,1,1,3,2,2] #학년 column 추가ㅇ
df
```

```python
df.groupby(['학교','학년']).mean(numeric_only=True) #학교, 학년별 평균
```

```python
df.groupby(['학년']).mean(numeric_only=True)
```

```python
df.groupby('학년').mean(numeric_only=True).sort_values('키')
```

```python
df.groupby('학년').mean(numeric_only=True).sort_values('키',ascending=False)
```

```python
df.groupby(['학교','학년']).sum()
```

```python
df.groupby('학교')['SW특기'].count() #학교로 그룹화 한뒤 SW특기 수
```

```python
df.groupby('학교')[['이름','SW특기']].count()
```

```python
school = df.groupby('학교')
school['학년'].value_counts()
```

```python
school['학년'].value_counts().loc['북산고'] #학교로 그룹화 한 뒤 북산고에 대해 학년별 학생수
```

```python
school['학년'].value_counts().loc['능남고']
```

```python
school['학년'].value_counts(normalize=True).loc['북산고'] #학생들의 수 데이터를 퍼센트로
```

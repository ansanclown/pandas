# 8. 막대그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus'] = False #마이너스 글자 꺠짐
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels,values)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels,values, color='r')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
colors = ['r','g','b']

plt.bar(labels,values, color=colors, alpha=0.5)
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels,values)
plt.ylim(175,195)
```

```python
plt.bar(labels,values, width=0.5)
```

```python
plt.bar(labels,values,width=0.3)
plt.xticks(rotation=45) #x축의 이름데이터 각도를 45도
plt.yticks(rotation=45) #y추그이 키 데이터 각도를 45도
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
ticks=['1번학생','2번학생','3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```

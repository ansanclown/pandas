# Matplotlib
다양한 형태의 그래프 통해서 데이터 시각화 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.show()
```

## 타이틀 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺽은선 그래프')
```

# 1. Linear Regression
공부시간에 따른 시험 점수

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1].values #처음부터 마지막 컬럼 직전까지 데이터 (독립변수)
y=dataset.iloc[:,-1].values #마지막 컬럼 데이터 (종속변수)
X,y
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression() #객체생성
reg.fit(X,y) #학습 (모델 생성)
```

```python
y_pred = reg.predict(X) # X에 대한 예측값
y_pred
```

```python
plt.scatter(X, y, color='blue') #산점도
plt.plot(X, y_pred, color='green') #선
plt.title('score by hours') #제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
print('9시간 공부했을 때 예상 점수 : ', reg.predict([[9]]))
```

```python
reg.coef_ #기울기 (m)
```

```python
reg.intercept_ #y절편 (b)
```

### 데이터 세트 분리

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0) #train:test=8:2분리
```

```python
X, len(X) #전체 데이터 X, 갯수
```

```python
X_train, len(X_train) #훈련 set X
```

```python
X_test, len(X_test)
```

```python
y, len(y)
```

```python
y_train, len(y_train) #훈련세트 y
```

```python
y_test, len(y_test) #테스트세트 y
```

### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train, y_train) #훈련세트로 훈련
```

### 데이터 시각화 (훈련세트) 

```python
plt.scatter(X_train, y_train, color='blue') #산점도
plt.plot(X_train, reg.predict(X_train), color='green') #선
plt.title('score by hours (trian data)') #제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color='blue') #산점도
plt.plot(X_train, reg.predict(X_train), color='green') #선
plt.title('score by hours (test data)') #제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델평가

```python
reg.score(X_test, y_test) #테스트 세트를 통한 모델 평가
```

```python
reg.score(X_train, y_train) #훈련세트를 통한 모델 평가
```

## 경사하강법(gradient descent)

max iter : 훈련 세트 반복 횟수 (epoch 횟수)  
eta0 : 학습률 (learning rate)

```python
from sklearn.linear_model import SGDRegressor #SGD : 확률적 강사 하강법
#지수표기법
#1e-3 : 0.001 (10^-3)
#1e+3 : 1000 (10^3)

#sr = SGDRegressor(max_iter=200, eta0=1e-4, random_state=0, verbose=1)
sr = SGDRegressor()
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue') #산점도
plt.plot(X_train, sr.predict(X_train), color='green') #선
plt.title('score by hours (train data SGD)') #제목
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
sr.coef_
```

```python
sr.intercept_
```

```python
sr.score(X_test, y_test)
```

```python
sr.score(X_train, y_train)
```

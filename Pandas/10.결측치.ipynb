## 데이터 제외하기 dropna

```python
import pandas as pd
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.dropna(inplace=True) #NaN 포함하는 데이터 삭제
```

```python
df
```

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
df
```

- axis : index or columns(row 지울지 column 지울지)  
- how : any or all (하나라도 있으면 지울지, 전체가 있으면 지울지 / 기본은 any)

```python
df.dropna(axis='index', how='any') #NaN이 하나라도 있으면  row 삭제
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') #column 데이터 전체가 NaN인 경우만  column 삭제
```
